```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df_global = pd.read_csv('F:\\Dropbox\\Data Science\\Nanodegree\\global_data.csv')
df_global
```

```python
df_global['10year_avg'] = df_global['avg_temp'].rolling(10).mean()
df_global
```

```python
import seaborn as sns
sns.lineplot(x = 'year', y ='10year_avg', data=df_global)
```

```python
df_city = pd.read_csv('F:\\Dropbox\\Data Science\\Nanodegree\\city_data.csv')
df_city.groupby('city')
df_seattle = df_city[df_city['city'] == 'Seattle']
df_seattle.head()
```

```python
df_seattle = df_seattle.reset_index()
df_seattle['10year_avg'] = df_seattle['avg_temp'].rolling(10).mean()
```

```python
df_seattle
```

```python
f, ax = plt.subplots(figsize=(8,6))
sns.lineplot(x = 'year', y ='10year_avg', data=df_seattle)
sns.lineplot(x = 'year', y ='10year_avg', data=df_global)
ax.legend(['seattle','global'],fontsize='x-large')
```

```python
df_global['5year_avg'] = df_global['avg_temp'].rolling(5).mean()
df_seattle['5year_avg'] = df_seattle['avg_temp'].rolling(5).mean()
f, ax = plt.subplots(figsize=(8,6))
sns.lineplot(x = 'year', y ='5year_avg', data=df_seattle)
sns.lineplot(x = 'year', y ='5year_avg', data=df_global)
ax.legend(['seattle','global'],fontsize='x-large')
```

```python
df_global['20year_avg'] = df_global['avg_temp'].rolling(20).mean()
df_seattle['20year_avg'] = df_seattle['avg_temp'].rolling(20).mean()
f, ax = plt.subplots(figsize=(8,6))
sns.lineplot(x = 'year', y ='20year_avg', data=df_seattle)
sns.lineplot(x = 'year', y ='20year_avg', data=df_global)
ax.legend(['seattle','global'],fontsize='x-large')
plt.xticks(np.arange(1760, 2020, 20))
plt.yticks(np.arange(7, 9.7, 0.2))
```
